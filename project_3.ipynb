```python
import pandas as pd
df = pd.read_csv('project-3.csv')
```

```python
missing = df.isnull().sum()
print(" Missing Values Per Column:")
print(missing)
```

```python
df.describe()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["Year"], df["Births"])
plt.title("Year vs Births")
plt.xlabel("Year")
plt.ylabel("Births")
plt.show()
```

```python
plt.figure(figsize=(8,4))
df.groupby("Year")["Births"].sum().plot(kind="line", marker="o")
plt.title("Births Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Total Births")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(6,6))
df.groupby("Country")["Births"].sum().plot(
    kind="pie",
    autopct="%1.1f%%"
)
plt.title("Births Distribution by Country")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(8,4))
df.groupby("Country")["Births"].sum().plot(kind="bar")
plt.title("Total Births by Country")
plt.xlabel("Country")
plt.ylabel("Total Births")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,4))
(df.groupby("Year")["Births"].sum() / (365*24*60*60)).plot(kind="bar")
plt.title("Births per Second by Year")
plt.xlabel("Year")
plt.ylabel("Births per Second")
plt.xticks(rotation=45)
plt.show()
```

```python
print("Total Births:", df["Births"].sum())
print("Highest Birth Country:", df.groupby("Country")["Births"].sum().idxmax())
print("Most Recent Year:", df["Year"].max())
```
